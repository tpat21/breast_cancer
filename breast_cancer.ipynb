```python
# Model Training
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target'], axis = 1)
y = df['target']
```

```python
x
```

```python
y = df['target']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state=5)
```

```python
x_train_shape = x_train.shape
x_test_shape = x_test.shape
y_train_shape = y_train.shape
y_test_shape = y_test.shape

print("The shape of x_train is: {}".format(x_train_shape))
print("The shape of x_test is: {}".format(x_test_shape))
print("The shape of y_train is: {}".format(y_train_shape))
print("The shape of y_test is: {}".format(y_test_shape))
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC(gamma='auto')
svc_model.fit(x_train, y_train)
```

```python
# EVALUATING THE MODEL
```

```python
y_predict = svc_model.predict(x_test)
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
true_pos = int(cm[0,[0]]) + int(cm[1,[1]])
type_one = int(cm[0,[1]])
type_two = int(cm[1,[0]])


print("There are {} correct, {} Type I errors, and {} Type II errors predictions".format(true_pos,type_one,type_two))
```

```python
print(classification_report(y_test, y_predict))
```

```python
# Imporving the model
```

```python
min_train = x_train.min()
min_train
```

```python
range_train = (x_train - min_train).max()
range_train
```

```python
x_train_scaled = (x_train - min_train)/range_train
x_train_scaled
```

```python
sns.scatterplot(x = x_train['mean area'],y = x_train['mean smoothness'],hue = y_train)
```

```python
min_test = x_test.min()
range_test = (x_test - min_test).max()
x_test_scaled = (x_test - min_test)/range_test
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC(gamma='auto')
svc_model.fit(x_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm,annot=True,fmt="d")
```

```python
true_pos = int(cm[0,[0]]) + int(cm[1,[1]])
type_one = int(cm[0,[1]])
type_two = int(cm[1,[0]])


print("There are {} correct, {} Type I errors, and {} Type II errors predictions".format(true_pos,type_one,type_two))
```

```python
print(classification_report(y_test,y_predict))
```

```python
# Optimizing Parameters
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100], 'gamma':[1,.1,0.1,.001], 'kernel':['rbf']}

grid = GridSearchCV(SVC(),param_grid, refit= True, verbose = 4)
grid.fit(x_train_scaled, y_train)
```

```python
optimized_param = grid.best_params_
print("Our model is most optimal when C = {} and when gamma = {}".format(optimized_param.get("C"),optimized_param.get("gamma")))
```

```python
grid_predict = grid.predict(x_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,grid_predict))
```
